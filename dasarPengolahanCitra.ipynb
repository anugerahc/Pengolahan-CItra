```python
#Fungsi untuk menjalankan Gambar
from google.colab.patches import cv2_imshow
import cv2 as cv

#Fungsi untuk menginput citra
def cvCitra(value, kode):
  gambar = cv.imread(value, kode) #angka 0 menunjukan gambar grayscale, jika 1 maka gambar berwarna

  cv2_imshow(gambar)
  cv.waitKey(0)

#Perintah Untuk Menampilkan Citra
print('Menampilkan Citra dengan openCV')
cvCitra('/content/src/citra1.jpeg', 1)
cvCitra('/content/src/citra2.jpeg', 0)
cvCitra('/content/src/citra3.jpeg', 1)
cvCitra('/content/src/citra4.jpeg', 1)
cvCitra('/content/src/citra5.jpeg', 0)

```


6. Operasi Aritmatika Terhadap Citra

```python
from google.colab.patches import cv2_imshow
import cv2 as cv

#Fungsi pengolahan Aritmatika Citra
def aritmatikaCitra(gambar1, gambar2, nama1, nama2):

  print('Operasi Aritmatika ' + nama1 + ' dan ' + nama2 + '\n')
  gambar1 = cv.imread(gambar1)
  gambar2 = cv.imread(gambar2)
  penjumlahan = cv.add(gambar1, gambar2)
  pengurangan = cv.subtract(gambar1, gambar2)
  perkalian = cv.multiply(gambar1, gambar2)
  pembagian = cv.divide(gambar1, gambar2)

  print('Penjumlahan Citra ' + nama1 + ' dan ' + nama2)
  cv2_imshow(penjumlahan)
  print()

  print('Pengurangan Citra ' + nama1 + ' dan ' + nama2)
  cv2_imshow(pengurangan)
  print()

  print('Perkalian Citra ' + nama1 + ' dan ' + nama2)
  cv2_imshow(perkalian)
  print()

  print('Pembagian Citra ' + nama1 + ' dan ' + nama2)
  cv2_imshow(pembagian)
  print()
  
  cv.waitKey(0)

#Fungsi pengolahan Logika Citra
def logikaCitra(gambar1, gambar2, nama1, nama2):

  print('Operasi Logika (Boolean) ' + nama1 + ' dan ' + nama2 + '\n')
  gambar1 = cv.imread(gambar1)
  gambar2 = cv.imread(gambar2)
  operasiNot = cv.bitwise_not(gambar1)
  operasiOr = cv.bitwise_or(gambar1, gambar2)
  operasiAnd = cv.bitwise_and(gambar1, gambar2)
  operasiXor = cv.bitwise_xor(gambar1, gambar2)


  print('Logika NOT ' + nama1 + ' dan ' + nama2)
  cv2_imshow(operasiNot)
  print()

  print('Logika OR ' + nama1 + ' dan ' + nama2)
  cv2_imshow(operasiOr)
  print()

  print('Logika AND ' + nama1 + ' dan ' + nama2)
  cv2_imshow(operasiAnd)
  print()

  print('Logika XOR ' + nama1 + ' dan ' + nama2)
  cv2_imshow(operasiXor)
  print()
  
  cv.waitKey(0)

  
#Pemanggilan Fungsi
aritmatikaCitra('/content/src/citra3.jpeg', '/content/src/citra5.jpeg','Citra3', 'Citra5')
logikaCitra('/content/src/citra6.jpeg', '/content/src/citra7.jpeg','Citra6', 'Citra7')
```

 7. Operasi Geometri Terhadap Citra



```python
from google.colab.patches import cv2_imshow
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

#Fungsi Penskalaan(Scaling)
def penskalaan(value):
  gambar = cv.imread(value)

  (height, width) = gambar.shape[:2]
  hasil = cv.resize(gambar, (int(width / 2), int(height /2)), interpolation = cv.INTER_CUBIC)

  print("Penskalaan")

  print('Sebelum')
  cv2_imshow(gambar)

  print()

  print('Sesudah')
  cv2_imshow(hasil)

  print()

#Fungsi Pergeseran(Translasi)
def pergeseran(value):
  gambar = cv.imread(value, 0)
  rows, cols = gambar.shape
  M = np.float32([[1, 0, 100], [0, 1, 10]])
  hasil = cv.warpAffine(gambar, M, (cols, rows))

  print('Pergeseran')

  print('Sebelum')
  cv2_imshow(gambar)
  print()

  print('Sesudah')
  cv2_imshow(hasil)
  print()

#Fungsi Perputaran(Rotasi)
def rotasi(value):
  gambar = cv.imread(value, 0)
  rotasi90Derajat = cv.rotate(gambar, cv.ROTATE_90_CLOCKWISE)

  print('Rotasi')
  print('Sebelum')
  cv2_imshow(gambar)
  print()

  print('Sesudah')
  cv2_imshow(rotasi90Derajat)

#Pemanggilan Fungsi
penskalaan('/content/src/citra4.jpeg')
pergeseran('/content/src/citra7.jpeg')
rotasi('/content/src/citra6.jpeg')
```
